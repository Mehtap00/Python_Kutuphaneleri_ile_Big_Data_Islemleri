```python
def divide_thread(data, c, y, min_similarity, thread_number):
    s_time = time.time()
    
    n = len(data)
    h = math.floor(n / 2 / thread_number)

    thread_list = []
    for t in range(0, thread_number):
        thread_list.append(None)

    with concurrent.futures.ThreadPoolExecutor() as executor:
        for t in range(0, thread_number):
            if n % 2 == 1:
                if t == 0:
                    ##thread_list[t]=threading.Thread(target=similarity_calculate,args=(t*h , (t*h)+h ,c, n-(t*h)-h-1 , n-(t*h)-1 , t , data,y,min_similarity))
                    thread_list[t] = executor.submit(similarity_calculate, t * h, (t * h) + h, c, n - (t * h) - h - 1,
                                                     n - (t * h) - 1, t, data, y, min_similarity)

                else:
                    if t == thread_number - 1:
                        ##thread_list[t]=threading.Thread(target=similarity_calculate,args=(t*h , math.floor(n/2) ,c, math.floor(n/2) , n-(t*h)-1 , t , data,y,min_similarity))
                        thread_list[t] = executor.submit(similarity_calculate, t * h, math.floor(n / 2), c,
                                                         math.floor(n / 2), n - (t * h) - 1, t, data, y, min_similarity)

                    else:
                        ##thread_list[t]=threading.Thread(target=similarity_calculate,args=(t*h , (t*h)+h ,c, n-(t*h)-h-1 , n-(t*h)-1 , t , data,y,min_similarity))
                        thread_list[t] = executor.submit(similarity_calculate, t * h, (t * h) + h, c,
                                                         n - (t * h) - h - 1, n - (t * h) - 1, t, data, y,
                                                         min_similarity)

            else:
                if t == 0:
                    ##thread_list[t]=threading.Thread(target=similarity_calculate,args=(t*h , (t*h)+h ,c, n-(t*h)-h , n-(t*h)-1 , t , data,y,min_similarity))
                    thread_list[t] = executor.submit(similarity_calculate, t * h, (t * h) + h, c, n - (t * h) - h,
                                                     n - (t * h) - 1, t, data, y, min_similarity)

                else:
                    if t == thread_number - 1:
                        ##thread_list[t]=threading.Thread(target=similarity_calculate,args=(t*h , math.floor(n/2) ,c, math.floor(n/2) , n-(t*h) , t ,data,y,min_similarity))
                        thread_list[t] = executor.submit(similarity_calculate, t * h, math.floor(n / 2), c,
                                                         math.floor(n / 2), n - (t * h), t, data, y, min_similarity)

                    else:
                        ##thread_list[t]=threading.Thread(target=similarity_calculate,args=(t*h , (t*h)+h ,c, n-(t*h)-h , n-(t*h) , t ,data,y,min_similarity))
                        thread_list[t] = executor.submit(similarity_calculate, t * h, (t * h) + h, c, n - (t * h) - h,
                                                         n - (t * h), t, data, y, min_similarity)

    print("--------Toplam Çalışma Zamanı:" + str(time.time() - s_time) + " seconds--------")
```

```python
def divide_thread2(c, index, y, min_similarity, thread_number):
    s_time = int(time.time())

    f = len(data)
    n = math.floor(f / thread_number)

    thread_list = []
    for t in range(0, thread_number):
        thread_list.append(None)

    for t in range(0, thread_number):

        if n == f / thread_number:
            ##print(str(t*n)+" "+str((t+1)*n))
            thread_list[t] = threading.Thread(target=similarity_calculate2,
                                              args=(c, index, t * n, (t + 1) * n, t, y, min_similarity))

        else:
            if t == thread_number - 1:
                ##print(str(t*n)+" "+str(((t+1)*n)+1))
                thread_list[t] = threading.Thread(target=similarity_calculate2,
                                                  args=(c, index, t * n, ((t + 1) * n) + 1, t, y, min_similarity))
            else:
                ##print(str(t*n)+" "+str((t+1)*n))
                thread_list[t] = threading.Thread(target=similarity_calculate2,
                                                  args=(c, index, t * n, (t + 1) * n, t, y, min_similarity))

    for t in range(0, thread_number):
        thread_list[t].start()

    for t in range(0, thread_number):
        thread_list[t].join()

    print("--------Toplam Çalışma Zamanı:" + str(int(time.time()) - s_time) + " seconds--------")
```
